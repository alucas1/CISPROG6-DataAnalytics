```python
import numpy as np
import pandas as pd
```

```python
#Open and read CSV file
df = pd.read_csv("podcasts.csv", encoding='windows-1252')
```

```python
#Creates a table showing all podcasts between a given range of downloads
lower = 1000000
upper = 5000000

range_pods = df[df['Downloads'].isin(range(lower, upper + 1))] 
range_pods
```

```python
#Outputs ancient era podcast in sorted order
ancient_pods_sorted = df[df['Era']=='Ancient'].sort_values(by ='Downloads')
ancient_pods_sorted
```

```python
#Outputs the most downloaded ancient era podcast's date of release
ancient_date = ancient_pods_sorted['Release date'].iloc[-1]
ancient_date
```
